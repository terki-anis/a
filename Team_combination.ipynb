**Le but de cette fonction est étant donné un certain nombre de joueurs dans un (ou plusieurs) postes elle toutes affiche les possibilités des tactiques avec cette condition sur les postes ainsi que les combinaisons de joueurs dans ce poste de chaque tactique**

Par exemple si on choisit le nombre de défenseurs= 5 on aura toutes les tactiques avec 5 défenseurs (5-3-3, 5-4-1, 5-2-3...) et pour chaque tactique on aura les combinaisons possibles des 11 joueurs parmi 15.

```python
def tactique(team,defense=None,milieu=None,attaque=None):
  if defense is not None:  #Vérifier si l'utilisateur spécicfie le nombre de défenseurs
    defense=[defense]
  else:
    defense=list(range(1,7))    #si le nombre n'est pas spécifié on prend toutes les valeurs possibles
                              #Il est a noter que nous n'avons pas pris 0 dans notre intervalle pour ne pas tomber sur des équipes avec 0 défenseurs
  if milieu is not None:
    milieu=[milieu]
  else:
    milieu=list(range(1,5))

  if attaque is not None:
    attaque=[attaque]
  else:
    attaque=list(range(1,4))

  #Maintenant qu'on a toutes les conditions de notre équipe nous allons iterer sur toutes les combinaisons possible de nos 11 joueurs   
  for comb in list(combinations(team['player'],11)):
    squad=team[team['player'].isin(list(comb))]
    #squad représente la liste de joueurs de chaque combinaisons ainsi que leurs poistions respectives

    goalkeeper=list(squad['position']).count('goalkeeper')
    defender= list(squad['position']).count('defender')      #On compte le nombre de défenseurs de cette combinaison
    midfielder = list(squad['position']).count('midfielder')
    striker = list(squad['position']).count('striker')

    if((goalkeeper==1) and ( defender in defense) and (midfielder in milieu) and (striker in attaque)):
      #si une combinaisons respecte toutes les conditions (1 seul gardien), et nombre de poste correspondants
      #on affiche les possibilités de tactiques ainsi que les combinaisons de joueurs qui leur correspondent
      print(defender,midfielder,striker)
      print(squad)
```
